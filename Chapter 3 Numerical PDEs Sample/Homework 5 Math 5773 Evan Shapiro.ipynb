Exercise 3.18 Let a = a(x,t,u) be a strictly positive>0,
function. Use energy arguments to show that any solution of the (possibly nonlinear) problem
$$
u_{t} = (a(x,t,u)u_{x})_{x} \quad for \quad x \in(0,1), t>0,\\
u(0,t) = u(1,t) = 0,\\
u(x,0)= f(x)\\
$$
satisfies the estimate
$$
\int_{0}^{1}u^2(x,t)dx \le \int_{0}^{1}f(x)^2dx \quad t>0
$$
Solution: Use the same arguments used to derive energy bounds for the normal heat equation.
$$
E(t) = \int_{0}^{1}u^2(x,t)dx\\
E'(t) = 2\int_{0}^{1}uu_{t}dx\\
E'(t) = 2\int_{0}^{1}u(a(x,t,u)u_{x})_{x}dx\\
$$
Use integration by parts to evaluate this integral.
$$
E'(t) = 2(ua(x,t,u)u_{x}|_{0}^{1} - \int u_{x}^2a(x,t,u)dx)\\
$$
By the given boundary conditions the first term is zero. a(x,t,u) has been defined to be strictly positive, so the second term 
$$
 - 2\int u_{x}^2a(x,t,u)dx
 $$
 is less than zero for all t>0. Therefore
 $$
 \int_{0}^{1}u^2(x,t)dx \le \int_{0}^{1}f(x)^2dx \quad for \quad t>0
 $$

Exercise 3.20 Consider the problem
$$
u_{t} = u_{xx} + u \quad for \quad x \in (0,1), t>0,\\
u(0,t) = u(1,t) = 0,\\
u(x,0) = f(x).
$$
Show that
$$
\frac{d}{dt}(e^{-2t}\int_{0}^{1}u^2(x,t)dx)\le 0
$$
and conclude that 
$$
\int_{0}^{1}u^2(x,t)dx \le e^{2t}\int_{0}^{1}f^2(x)dx
$$
Solution:
Let $E(t) = e^{-2t}\int_{0}^{1}u^2(x,t)dx$
$$
E'(t) = \frac{d}{dt}(e^{-2t}\int_{0}^{1}u^2(x,t)dx) = -2e^{-2t}\int_{0}^{1}u^2(x,t)dx+2e^{-2t}\int_{0}^{1}u(x,t)u_{t}(x,t)dx\\
= 2e^{-2t}(\int_{0}^{1}u(u_{xx}+u)dx - \int_{0}^{1}u^2dx)\\
=2e^{-2t}(\int_{0}^{1}uu_{xx}dx + \int_{0}^{1}u^2dx- \int_{0}^{1}u^2dx)\\
=2e^{-2t}\int_{0}^{1}uu_{xx}dx
$$
With the given boundary conditions, the integral was shown to be less than or equal to zero in section 3.7. Since the energy is decreasing with respect to time we can say
$$
 e^{-2t}\int_{0}^{1}u^2(x,t)dx \le  e^{-2t_{0}}\int_{0}^{1}u^2(x,t_{0})dx\\
  e^{-2t}\int_{0}^{1}u^2(x,t)dx \le \int_{0}^{1}f(x)^2dx\\
\int_{0}^{1}u^2(x,t)dx \le e^{2t}\int_{0}^{1}f(x)^2dx
$$

Use this estimate to bound the difference between the two solutions in terms of differences between the initial functions. Does the problem have a unique solution for each initial function f(x)?
$$
$$
Solution: Assume that $u_{1}$ and $u_{2}$ are solution to the given initial boundary value problem, with inital functions $f_{1}(x)$ and $f_{2}(x)$, respectively. Let $w = u_{1} -u_{2}$. As per exercise 3.19, $w$ satisfies the boundary conditions of the original problem, with a new initial function $w(x,0) = f_{1}(x) - f_{2}(x)$. The new problem is given by
$$
w_{t} = w_{xx} + w,
$$
which has the same form as the original problem.
Therefore, the new problems energy has the same bound as the original problem
$$
\int_{0}^{1}(u(x,t)_{1} - u(x,t)_{2})^2dx = \int_{0}^{1}w(x,t)^2dx \le e^{2t}\int_{0}^{1}(f(x)_{1} - f(x)_{2})^2dx
$$
This indicates that this nonlinear problem is unstable with respect to small perturbations in the initial functions, and that over time the differences between two solutions, $u_{1}$ and $u_{2}$, grows exponentially. If the two initial functions  are equivalent, $f_{1}(x)=f(x)_{2}$, then the difference is zero and the solution is uniquely determined.
